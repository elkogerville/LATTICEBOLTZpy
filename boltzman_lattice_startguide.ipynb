# LATTICEBOLTZpy
this python program performs lattice boltzman hydrodynamical simulation for given boundaries

```python
from LATTICEBOLTZpy import *
import numpy as np
import matplotlib.pyplot as plt
```

### define simulation parameters
we now define a meshgrid of particles that define the simulation space

```python
# define simulation space
Nx = 500
Ny = 400
# create meshgrid
X, Y = np.meshgrid(range(Nx), range(Ny))
# create boolean array defining a city block
city_block = (((X>50) & (X<100) & (Y>50) & (Y<100)) | ((X>130) & (X<260) & (Y>125) & (Y<155))
           | ((X>130) & (X<180) & (Y>50) & (Y<100)) | ((X>210) & (X<260) & (Y>50) & (Y<100))
           | ((X>290) & (X<340) & (Y>50) & (Y<100)) | ((X>50) & (X<100) & (Y>300) & (Y<350)) 
           | ((X>130) & (X<180) & (Y>300) & (Y<350)) | ((X>210) & (X<260) & (Y>300) & (Y<350))
           | ((X>290) & (X<340) & (Y>300) & (Y<350)) | ((X>290) & (X<340) & (Y>225) & (Y<275))
            | ((X>290) & (X<340) & (Y>125) & (Y<175)) | ((X>50) & (X<100) & (Y>225) & (Y<275))
            | ((X>50) & (X<100) & (Y>125) & (Y<175)) | ((X>130) & (X<260) & (Y>240) & (Y<275))
            | ((X>150) & (X<240) & (Y>170) & (Y<175)) | ((X>150) & (X<240) & (Y>190) & (Y<195)) 
             | ((X>150) & (X<240) & (Y>215) & (Y<220)) 
           )
```

```python
# display the city block, viewed in bird's eye view
fig = plt.figure(figsize = (7,7))
plt.imshow(city_block)
```

### running the simulation will plot the timesteps

```python
boltzman_lattice(objects = city_block, Nx = Nx, Ny = Ny, v0 = 2.3, timesteps = 1000)
```
